```python
import spacy,random,requests
import re

nlpEn = spacy.load("en_core_web_md")
nlpEs = spacy.load("es_core_news_md")
nlpPt = spacy.load("pt_core_news_md")
nlpIt = spacy.load("it_core_news_md")
nlpFr = spacy.load("fr_core_news_md")
```

```python
#module mlconjug3 permet de conjuguer facilement en langue romane
import mlconjug3

conj_fr = mlconjug3.Conjugator(language='fr')
conj_en = mlconjug3.Conjugator(language='en')
conj_es = mlconjug3.Conjugator(language='es')
conj_it = mlconjug3.Conjugator(language='it')
conj_pt = mlconjug3.Conjugator(language='pt')
```

```python
import pandas as pd
```

```python
data = pd.read_csv("Langue Romane.csv")
data2 = pd.read_csv("Langue Romane - Phrase.csv")
```

```python
def Flashcard(lang, target):
    size = data[lang].size
    rand = random.randint(0, size)
    return data[[lang, target]]

Flashcard("French", "Spanish")
```

```python
def select_conj(lang):
    if lang == "French":
        conj = conj_fr
    if lang == "Spanish":
        conj = conj_es
    if lang == "Portuguese":
        conj = conj_pt
    if lang == "Italiano":
        conj = conj_it
    return conj
```

```python
def test(verb, lang, temps):
    conj = select_conj(lang)
    test_verb = conj.conjugate(verb)
    all_conjugated_forms = test_verb.iterate()
    for i in all_conjugated_forms:
        print(i)
    
test("hablar", "Spanish", "Indicativo presente")
```

```python
# 1 verbe 1 langue 1 temps donne la conjugaison pour chaque personnes
#display simplement la conj à toutes les personnes

def conju1(verb, lang, mode, temps):
    conj = select_conj(lang)
    test_verb = conj.conjugate(verb)
    all_conjugated_forms = test_verb.iterate()
    df = pd.DataFrame(all_conjugated_forms, columns = ['MODE', "TEMPS", "PERSONNE", "CONJUGAISON"])
    df = df[['MODE', "TEMPS", "PERSONNE", "CONJUGAISON"]][(df["TEMPS"] == temps) & (df["MODE"] == mode)]

    display(df)
conju1("parler", "French", "Indicatif", "Présent")
conju1("hablar", "Spanish", "Indicativo", "Indicativo presente")
#conju1("falar", "Portuguese", "Indicativo", "Indicativo presente")
#conju1("palare", "Italiano", "Indicativo", "Indicativo presente")
```
